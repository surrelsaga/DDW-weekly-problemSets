**HW2.** *Max-Heapify:* Recall the algorithm for max-heapify in restoring the *max-heap property* of a binary heap tree. In the previous implementation, we used iteration. Implement the same algorithm using recursion. 

Reorder the sequence of the code to implement max heapify function.

````python
# copy these functions from you cohort problem set

def left_of(index: int) -> int:
    ###
    return 2 * index + 1
    ###

def right_of(index: int) -> int:
    ###
    return (index + 1) * 2
    ###

def max_child(array: list[int|float], index: int, heap_size: int) -> int:
    ###
    if right_of(index) >= heap_size: #out of index
        return left_of(index)
    if array[ left_of(index) ] > array[ right_of(index) ]:
        return left_of(index)
    else:
        return right_of(index)
        
    ###

````

````python
def max_heapify(array: list, index: int, heap_size: int) -> None:
    ###
    # RECURSION
    
    # Base case: no left child -> node is a leaf
    if left_of(index) >= heap_size:
        return
    
    max_child_idx = max_child(array, index, heap_size)
    
    if array[max_child_idx] > array[index]:
        # swap
        array[max_child_idx], array[index] = array[index], array[max_child_idx]
        
        # continue with the node that has max_child_i
        max_heapify(array, max_child_idx, heap_size)
    
    ###
````

````python
result: list[int] = [16, 4, 10, 14, 7, 9, 3, 2, 8, 1]
max_heapify(result, 1, len(result))
print(result)
assert result == [16, 14, 10, 8, 7, 9, 3, 2, 4, 1]
result: list[int] = [4, 1, 10, 14, 16, 9, 3, 2, 8, 7]
max_heapify(result, 1, len(result))
print(result)
assert result == [4, 16, 10, 14, 7, 9, 3, 2, 8, 1]
````
