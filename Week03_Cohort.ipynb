```python
# recursion
def factorial_recursion(n: int) -> int:
    if n == 0 or n == 1:
        return 1
    
    return n * factorial_recursion(n - 1)
```

**CS2.** *Helper Function:* Write a function `palindrome(s)`to check if the string `s` is a [Palindrome](https://en.wikipedia.org/wiki/Palindrome). To do this, write another function `is_palindrome(s, left, right)` where `left` and `right` are indices wherewith to check the substrings in `s`. Use recursion instead of iteration in this problem.


```python
def palindrome(s: str) -> bool:
    return is_palindrome(s, 0, len(s) - 1)

def is_palindrome(s: str, left: int, right: int) -> bool:
    # check the leftmost, rightmost pair to see if they are the same
    # result = True
    # for i in range(len(s) // 2):
    #     if s[i] != s[len(s) - 1]:
    #         result = False
    #         break
    
    # return result
    
    # recursion
    # base case:
    # 2 pointers: (this is the index that is pointed)
    # 1 point starting from the leftmost moving rightwards
    # 1 point starting from the rightmost leftwards
    
    # left == right, return true (odd string)
    # left == right - 1, return (s[left] == s[right]) (right string)
    # => compact: left >= right (left = right with odd string, left > right with even string)
    
    # recursive case:
    # is_palindrome(s, left, right) = ( s[left] == s[right] ) and is_palindrome(s, left+1, right-1)
    if left >= right:
        return True
    
    return (s[left] == s[right]) and is_palindrome(s, left + 1, right - 1)
    
    
```

```python
assert not palindrome("moon") 
assert palindrome("noon") 
assert palindrome("a a") 
assert palindrome("ada") 
assert not palindrome("ad a") 
```

```python
result: list[str] = move_disks(3, "A", "B", "C")
print(result)
assert result == ["Move disk 1 from A to B.", "Move disk 2 from A to C.", "Move disk 1 from B to C.", "Move disk 3 from A to B.", "Move disk 1 from C to A.", "Move disk 2 from C to B.", "Move disk 1 from A to B."]

result: list[str] = move_disks(4, "A", "B", "C")
print(result)
assert result == ["Move disk 1 from A to C.", "Move disk 2 from A to B.", "Move disk 1 from C to B.", "Move disk 3 from A to C.", "Move disk 1 from B to A.", "Move disk 2 from B to C.", "Move disk 1 from A to C.", "Move disk 4 from A to B.", "Move disk 1 from C to B.", "Move disk 2 from C to A.", "Move disk 1 from B to A.", "Move disk 3 from C to B.", "Move disk 1 from A to C.", "Move disk 2 from A to B.", "Move disk 1 from C to B."]
```
