The function should return `True` if the absolute difference between the **square** of the guess number and the actual number is less than the given tolerance. Otherwise, it should return `False`. 

Note:
- The input argument for tolerance should be optional. When it is not provided, it should use 0.001 as the default value for the tolerance.
- Use the previous definition of `absolute` and `square` in your implementation.

````python
# write your function definition for is_good_enough here

def is_good_enough(guess, actual, tolerance=0.001):
    difference = absolute( square(guess) - actual )
    if difference < tolerance:
        return True
    
    return False
````

Q7. Implement factorial problem using iteration. The function should takes in an Integer input and returns and Integer output which is the factorial of the input. Recall that:

$n! = n \times (n-1) \times (n-2) \times \ldots \times 2 \times 1$

You should consider the case when $n$ is zero and one as well. 

````python
# iteration
def factorial_iteration(n: int) -> int:
    result: int = 1
    
    if n == 0:
        return 1
    
    for i in range(1, n + 1):
        result *= i
        
    return result
````

````python
assert factorial_iteration(0) == 1
assert factorial_iteration(1) == 1
assert factorial_iteration(5) == 120
assert factorial_iteration(7) == 5040
assert factorial_iteration(11) == 39916800
````

$$Fib(n) = \left\{\begin{matrix}0 & \text{if } n = 0 \\ 1 & \text{if } n=1 \\ Fib(n-1) + Fib(n-2) & \text{otherwise}\end{matrix}\right.$$


````python
def fibonacci(n: int) -> int:
    if n <= 1:
        return n
    
    sum = 0
    # Because the start is 0, 1, 1, .. (start from here)
    left = 0
    right = 1
    for i in range(0, n):
        temp = right
        right = left + right
        left = temp
    
    return left
````

Q10. Create a function called `pascal_triangle(n)` that takes in an integer an returns a list of lists giving a Pascal triangle at height n. See article on [Pascal Triangle](https://en.wikipedia.org/wiki/Pascal%27s_triangle).

For example,
```
print(pascal_triangle(5))
```

has the following output.

```
[[1], [1, 1], [1, 2, 1], [1, 3, 3, 1], [1, 4, 6, 4,1]]
```

````python
def pascal_triangle(n: int) -> list[list[int]]:
    output: list[list[int]] = []
    for i in range(n):
        row = [1]
        if output:
            prev = output[-1]
            for j in range(len(prev) - 1):
                row.append(prev[j] + prev[j + 1])
            row.append(1)
        output.append(row)
    return output
````

````python
assert pascal_triangle(5) == [[1], [1, 1], [1, 2, 1], [1, 3, 3, 1], [1, 4, 6, 4, 1]]
````

Q11. Write a function named `find_average` that takes in a list of lists as an input. Each sublist contains numbers. The function returns a list of the averages of each sublist, and the overall average. If the sublist is empty, take the average to 0.0.

For example, if the input list is `[[3, 4], [5, 6, 7], [-1, 2, 3]]`, the program returns the list `[3.5, 6.0, 1.333]`, and the overall average 3.625, calculated by summing all the numbers in all the sublists and dividing this total sum by the total count of all the numbers.

```
>>> ans = find_average([[3,4],[5,6,7],[-1,2,8]])
>>> print(ans)
([3.5, 6.0, 3.0], 4.25)

>>> ans = find_average([[13.13,1.1,1.1],[],[1,1,0.67]])
>>> print(ans)
([5.11, 0.0, 0.89], 3.0)

>>> ans = find_average([[3.6],[1,2,3],[1,1,1]])
>>> print(ans)
([3.6, 2.0, 1.0], 1.8)

>>> ans = find_average([[2,3,4],[2,6,7],[10,5,15]])
>>> print(ans)
([3.0, 5.0, 10.0], 6.0)
```

````python
def find_average(lst: list[list[int|float]]) -> tuple[list[float], float]:
    output = []
    totalSum = 0
    total_count = 0
    for list in lst:
        if len(list) == 0:
            output.append(0.0)
            continue
        
        sum = 0
        for number in list:
            sum += number
            totalSum += number
            total_count += 1
        
        avg = sum / len(list)
        output.append(avg)
        
    
    return ( output, totalSum / total_count )
````

Q12. Use of a nested list in Python allows you to implement a data structure as a 2-dimensional matrix along with various matrix operations. Write a function named transpose_matrix, which takes a n x m integer matrix (i.e. a list with n items each of which is a list of m integer items) as an argument, and returns its transposed matrix. For example:

```
>>> a = [[1,2,3], [4,5,6], [7,8,9]]
>>> transpose_matrix(a)
[[1,4,7], [2,5,8], [3,6,9]]
>>> a = [[1,2,3], [4,5,6]]
>>> transpose_matrix(a)
[[1,4], [2,5], [3,6]]
```

Use a nested for-loop (i.e. a for loop inside a for loop) appropriately to implement this. You are not allowed to use any built-in transpose function from any libraries.

````python
from typing import Any 

def transpose_matrix(m: list[list[Any]]) -> list[list[Any]]:
    output = []
    
    # Since every list has the same length, we get that length first
    length = len(m[0])
    
    for i in range(length):
        list_to_add = [lst[i] for lst in m]
        
        output.append(list_to_add)
        
    return output
        
        
````

````python
assert transpose_matrix([[1,2,3],[4,5,6], [7, 8, 9]]) == [[1,4, 7],[2,5, 8],[3,6, 9]]
assert transpose_matrix([[1,2,3],[4,5,6]]) == [[1,4],[2,5],[3,6]]
````

Rewrite `bubble_up` function in your Cohort problem so that it can be used to implement [Bubble Sort version 3 in the Notes](https://data-driven-world.github.io/2023/notes/sorting-algorithms/bubble-sort/#bubble-sort-v3). To do this, the function should return the boolean value whether there is any swap or not.

````python
from typing import Optional

def bubble_up(array: list[int|float], end: Optional[int]=None) -> bool:
    if end is None:
        end = len(array)
    
    isSwap = False
    
    # n - 1 times to get n - 1 numbers in correct position
    for i in range(1, end):
        if array[i] < array[i - 1]:
            array[i], array[i - 1] = array[i - 1], array[i]
            isSwap = True
    
    return isSwap
    
````

````python
array: list[int|float] = [1, 100, 2, 3, 4, 5]
swap: bool = bubble_up(array)
assert array == [1, 2, 3, 4, 5, 100]
assert swap

array: list[int|float] = [1, 100, 2, 3, 4, 5]
swap = bubble_up(array, 5)
assert array == [1, 2, 3, 4, 100, 5]
assert swap


array: list[int|float] = [1, 2, 3, 4, 5]
swap = bubble_up(array)
assert not swap
````

Q13. Rewrite `bubble_sort` using [version 3](https://data-driven-world.github.io/2023/notes/sorting-algorithms/bubble-sort/#bubble-sort-v3) in the Notes with the `bubble_up` function written in the previous question.

````python
def bubble_sort(array: list[int|float]) -> None:
    n = len(array)
    
    # outer loop
    for i in range(n):
        swapped = bubble_up(array, n - i)
        
        # if there is no swap, the array is already swapped
        if not swapped:
            break
    
        
````

````python
array: list[int|float] = [10,3,8,47,1,0,-39,8,4,7,6,-5]
bubble_sort(array)
print(array)
assert array == [-39, -5, 0, 1, 3, 4, 6, 7, 8, 8, 10, 47]
````
