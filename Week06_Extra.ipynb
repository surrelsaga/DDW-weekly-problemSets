````python
def process_line(line: str) -> dict[str, int]:
    word_count: dict[str, int] = {}
    for word in line.split():
        if word in word_count:
            word_count[word] += 1
        else:
            word_count[word] = 1
    return word_count
````

Q2. Write a function called `add_daily_temp()` that takes in a dictionary of the average daily temperature, the day, and the value to update the day's average temperature. The dictionary should hold the average daily temperature for each day of the month. The function should add the temperature to the dictionary only if the dictionary does not already contain a temperature for that day. The function should return the resulting dictionary, whether or not it is updated.

````python
import copy

def add_daily_temp(avg_temp_dict: dict[int, float], day_index: int, avg_temp: float) -> dict[int, float]:
    output = copy.deepcopy(avg_temp_dict)
    
    if day_index >= 1 and day_index <= 31:
        if day_index not in output.keys():
            output[day_index] = avg_temp
    
    return output
````

````python
daily_temp: dict[int, float] = {}
output: dict[int, float] = add_daily_temp(daily_temp, 1, 30)
assert output == {1: 30}

output: dict[int, float] = add_daily_temp(daily_temp, 0, 30)
print(output)
assert output == {}
output: dict[int, float] = add_daily_temp(daily_temp, 32, 30)
assert output == {}

daily_temp : dict[int, float] = {1: 30}
output: dict[int, float] = add_daily_temp(daily_temp, 1, 32)
assert output == {1: 30}

output: dict[int, float] = add_daily_temp(daily_temp, 2, 32)
assert output == {1: 30, 2: 32}
````

Q3. Write a function `get_days_temp_range()` which takes in a dictionary containing the average daily temperature for each day in a month, a minimum temperature and a maximum temperature. The function returns a list of the days in which the average temperature is between the minimum and the maximum temperature, both inclusive. The output should be sorted.

````python
def get_days_temp_range(avg_temp_dict: dict[int, Number], min_temp: Number, max_temp: Number) -> list[int]:
    output: list[int] = []
    
    for day, temp in avg_temp_dict.items():
        if temp >= min_temp and temp <= max_temp:
            output.append(day)
            
            
    return output
````

````python
avg_temp: dict[int, Number] = {1: 30, 2: 32, 3: 31, 4: 29, 5: 31, 6: 32, 7: 33, 8: 34}

output: list[int] = get_days_temp_range(avg_temp, 30, 32)
print(output)
assert output == [1, 2, 3, 5, 6]
output: list[int] = get_days_temp_range(avg_temp, 31, 32)
assert output == [2, 3, 5, 6]
````

Q4. Write a function `get_source(graph, vert)` which returns a list of all source Vertices pointing to `vert` in the `graph`. For example, Vertex "C" has the following source Vertices: `["A", "B", "D", "F"]`. Return a `None` object if there are none.

````python
def get_source(graph: dict[str, list[str]], vert: str) -> Optional[Iterable[str]]:
    output: list[str] = []
    for vrtx, neighborlist in graph.items():
        if vert in neighborlist and vrtx not in output:
            output.append(vrtx)
    return output if len(output) > 0 else None
````

Q5. Write a function that takes in a graph representation in a dictionary and convert it to an object-oriented representation using the `Graph` class. Refer to your cohort problem sets for the `Graph` and `Vertex` class definition. 

The function takes in a dictionary that has the following format:
```python
graph = {'vertex1': [neighbour1, neighbour2, ...],
         'vertex2': [neighbour1, ...]}
```

Notes:
- The graph is represented as and adjecancy list. 
- The key of the dictionary is the `id` of the vertex in the graph. 
- The value of the dictionary is a list of neighbours of that vertex.
- The element in the list of neighbours is the `id` of the neighbouring vertices. 
- The data type for `id` is a string. 

Example:
```python
graph = {"A": ["B", "C"], 
         "B": ["C", "D"],
         "C": ["D"],
         "D": ["C"], 
         "E": ["F"],
         "F": ["C"]}
```

````python
class Vertex:
    def __init__(self, id_: str="") -> None:
        self.id_: str = id_
        self.neighbours: dict[Vertex, Number] = {}
    
    def add_neighbour(self, nbr_vertex: Vertex, weight: Number=0) -> None:
        self.neighbours[nbr_vertex] = weight
    
    def get_neighbours_list(self) -> list[Vertex]:
        # .keys() does not return a list type
        return list( self.neighbours.keys() )
    
    def get_weight(self, neighbour: Vertex) -> Optional[Number]:
        # return self.neighbours[neighbour]
        # if the neighbor is not in dict, then return None
        return self.neighbours.get(neighbour)
    
    def __eq__(self, other) -> bool:
        return self.id_ == other.id_
    
    def __lt__(self, other) -> bool:
        return self.id_ < other.id_
    
    def __hash__(self) -> int:
        return hash(self.id_)
    
    def __str__(self) -> str:
        output_str = f"Vertex {self.id_} is connected to: "
        neighbors_list = self.get_neighbours_list()
        
        for i in range( len(neighbors_list) ):
            neighbor_id = neighbors_list[i].id_
            
            # if the neighbor is the last neighbor in the list
            if (i == len(neighbors_list) - 1):
                str_to_add = f"{neighbor_id}"
            # else
            else:
                str_to_add = f"{neighbor_id}, "
            
            output_str += str_to_add
            
        return output_str
        
````

````python
class Graph:
    def __init__(self) -> None:
        self.vertices: dict[str, Vertex] = {}
    
    @property
    def num_vertices(self):
        return len( list(self.vertices.keys()) )
        
    def _create_vertex(self, id_: str) -> Vertex:
        return Vertex(id_)
    
    def add_vertex(self, id_: str) -> None:
        # create the vertex
        vertex_object = self._create_vertex(id_)
        # add to the dict
        self.vertices[id_] = vertex_object
    
    def get_vertex(self, id_: str) -> Optional[Vertex]:
        return self.vertices.get(id_)
    
    def add_edge(self, start_v: str, end_v: str, weight: Number=0) -> None:
        # Create vertices if they don't exist
        if start_v not in self.vertices:
            self.add_vertex(start_v)
        if end_v not in self.vertices:
            self.add_vertex(end_v)
        
        # Get the start vertex and add the end vertex as a neighbor
        start_vertex = self.vertices[start_v]
        end_vertex = self.vertices[end_v]
        start_vertex.add_neighbour(end_vertex, weight)
    
    # check if the vertex is inside the graph
    def __contains__(self, id_):
        return id_ in self.vertices.keys()
    
    def get_neighbours_ids(self, id_: str) -> list[str]:
        target_vertex = self.vertices[id_]
        
        # if there is no such vertex, return an empty list
        if id_ not in self.vertices:
            return []
        else:
            neighbors = [n.id_ for n in target_vertex.get_neighbours_list()]
            return neighbors
    
    def __iter__(self):
        for k,v in self.vertices.items():
            yield v
````

````python
def create_graph_object(g: dict[str, list[str]]) -> Graph:
    output: Graph = Graph()
    for start_v, neighbours in g.items():
        if start_v not in output.vertices:
            output.add_vertex(start_v)
        for end_v in neighbours:
            output.add_edge(start_v, end_v)
    return output
````

````python
graph: dict[str, list[str]] = {"A": ["B", "C"], 
         "B": ["C", "D"],
         "C": ["D"],
         "D": ["C"], 
         "E": ["F"],
         "F": ["C"]}

output: Graph = create_graph_object(graph)
assert output.num_vertices == 6
assert sorted([x.id_ for x in output]) == ["A", "B", "C", "D", "E", "F"]
assert sorted(output.get_neighbours_ids("A")) == ["B", "C"]
assert sorted(output.get_neighbours_ids("C")) == ["D"]
assert sorted(output.get_neighbours_ids("E")) == ["F"]
````

Q7. Write a function `is_cycle()`  that takes in a graph and returns a boolean. The function checks if the graph has a cycle. The way the function checks if a graph has a cycle is by traversing the graph using depth-first search. 

The algorithm keeps track visited node using a list `visited` and the path that the DFS is currently exploring using another list `rec_stack`. The list `rec_stack` represent the vertices in the current recursive stack. 

We know there is a cycle when there is a back edge. This means when our DFS algorithm is exploring the neighbours of the current vertex, it finds another vertex that has been visited and it is one of the vertices in the recursive stack list. Once it finds, the function should return `True` and exit all the function calls.

The difference between `visited` and `rec_stack` is that `visited` keeps track all the vertices that has been visited. On the other hand `rec_stack` removes those vertices that has been fully explored from the list. This means that `visited` list will continue to grow and adding all the vertices that have been visited. On the other hand, after you return from calling DFS for the neighbouring vertices of the current vertex, you should remove the current vertex from `rec_stack`.  

````python
def is_cycle(graph: dict[str, list[str]]) -> bool:
    visited: list[str] = []
    rec_stack: list[str] = []
    # initialize result to False
    result: Optional[bool] = False
    
    for k in graph:
        if k not in visited:
            # call DFS and store the output to result
            result = dfs(graph, k, visited, rec_stack)
            
            if result:
                return True
    return False

def dfs(graph: dict[str, list[str]], start: str, visited: list[str], rec_stack: list[str]) -> Optional[bool]:
    # add the start node to visited and rec_stack
    ###
    visited.append(start)
    rec_stack.append(start)
    ###

    # get the neighbours of the start node
    neighbours: Optional[list[str]] = graph[start]
    
    for neighbour in neighbours:
        if neighbour not in visited:
            # call DFS recursively
            result: Optional[bool] = dfs(graph, neighbour, visited, rec_stack)
            if result:
                return result
        elif neighbour in rec_stack:
            # if the neighbour is in rec_stack, then there is a cycle
            return True
    # remove the start node from rec_stack
    rec_stack.remove(start)

    return False

````

````python
graph_with_cycle: dict[str, list[str]] = {
    "a": ["b"],
    "b": ["c", "d"],
    "c": ["a"],
    "d": []
}

assert is_cycle(graph_with_cycle)
````

````python
graph_without_cycle: dict[str, list[str]] = {
    "a": ["b"],
    "b": ["c"],
    "c": ["d"],
    "d": []
} 

assert not is_cycle(graph_without_cycle)
````

Q8. One application of DFS is deadlock detection. In a computer system, a deadlock happens if there is a cycle in the resource allocation graph. 

![](https://www.dropbox.com/scl/fi/oyofo8xka1xrwj9obvaj7/resource_allocation_graph.png?rlkey=12y8z6247v82ldduyc4mrl1qg&raw=1)

Write a function that takes in a string describing a resource allocation graph in the following format and return a dictionary representing the directed graph.

```
R1 P1
P1 R3
P1 R2
R2 P2
P2 R1
P2 R4
```

It should output the following graph:
```
graph = {'R1': ['P1'],
         'P1': ['R3', 'R2'],
         'R2': ['P2'],
         'P2': ['R1', 'R4'],
         'R3': [],
         'R4': []}
```

````python
def generate_resource_allocation(input_str: str) -> dict[str, list[str]]:
    lines: list[str] = input_str.split("\n")
    resource_allocation: dict[str, list[str]] = {}
    
    for line in lines:
        if len(line) > 0:
            [first, second] = line.split()
            if first not in resource_allocation:
                resource_allocation[first] = [second]
            elif second not in resource_allocation[first]:
                resource_allocation[first].append(second)
                
            if second not in resource_allocation:
                resource_allocation[second] = []
    
    return resource_allocation
````

````python
allocation_data: str = """
R1 P1
P1 R3
P1 R2
R2 P2
P2 R1
P2 R4
"""

output: dict[str, list[str]] = generate_resource_allocation(allocation_data)
print(output)
assert output == {'R1': ['P1'],
         'P1': ['R3', 'R2'],
         'R2': ['P2'],
         'P2': ['R1', 'R4'],
         'R3': [],
         'R4': []}
````
