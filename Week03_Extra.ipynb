Q1. Implement a function `sum_arr(array)` to sum the element of a list. This function make use of another recursive function `sum_iter(array, index, sum)`. The second function uses recursion to iteratively sum the element of a list given by the index and keeps track of the sum using the third argument.

```python
from typing import TypeAlias
ListNumber: TypeAlias = list[int|float]
Number: TypeAlias = int|float
```

```python
def sum_arr(array: ListNumber) -> Number:
    return sum_iter(array, 0, 0)

def sum_iter(array: ListNumber, index: int, total: Number) -> Number:
    if index == len(array):
        return total
    else:
        total += array[index]
        return sum_iter(array, index + 1, total)
```

```python
assert sum_arr([1, 2, 3, 4, 5]) == 15
```

Q2. Write a function `factorial(n)` which calls a second function `fact_iter(product, counter, max_count)`. The second function implements iteration using recursive call to compute the factorial n. 

Note: In what way is this function different from that in your Cohort problem?

```python
def factorial(n: int) -> int:
    return fact_iter(1, 1, n)

def fact_iter(product: int, counter: int, max_count: int)-> int:
    if counter > max_count:
        return product
    else:
        product *= counter
        return fact_iter(product, counter + 1, max_count)
```

```python
print(factorial(5))
assert factorial(5) == 120
```

Q5. Write a function to compute exponentiation $b^n$. Exponentiation can be defined recursively using the following.

$$b^n = b \times b^{n-1}$$
$$b^0 = 1$$

```python
def expt(b: Number, n: int) -> Number:
    if n == 0:
        return 1
    if n == 1:
        return b
    
    return b * expt(b, n - 1)
```

```python
assert expt(2, 3) == 8
assert expt(3, 3) == 27
```

Q6. We can speed up exponentiation if $n$ are powers of 2. We can reduce the number of multiplication in this way.

$$b^2 = b \times b$$
$$b^4 = b^2 \times b^2$$
$$b^8 = b^4 \times b^4$$

In this way, when $n$ is even, we have
$$b^n = (b^{n/2})^2$$

and when $n$ is odd, we have
$$b^n = b\times b^{n-1}$$

Note: Use the previously defined `square()` function instead of applying `**` operator in Python.


```python
def square(x: Number) -> Number:
    return x * x
```

```python
def fast_expt(b: Number, n: int) -> Number:
    if n == 0:
        return 1
    if n == 1:
        return b
    if n % 2 == 0:
        return square(fast_expt(b, n // 2))
    else:
        return b * fast_expt(b, n - 1)
```

```python
assert fast_expt(2, 3) == 8
assert fast_expt(3, 3) == 27
assert fast_expt(2, 10) == 1024
```

Q7. Write a function `bin_str(n)` which converts an integer into a binary string representation. There are two base cases: when `n=0`, it should return `'0'`, and when `n=1`, it should return `'1'` as the string representation. For numbers larger than 1, we can do it recursively using

`concatenate(bin_str(n//2) , bin_str(n%2))`
where `concatenate` joins the two binary strings together. Do not use the built-in `bin` function!!

```python
def bin_str(n: int) -> str:
    # Base case
    if n == 0:
        return '0'
    if n == 1:
        return '1'
    
    # Recursive case
    # convert from decimal to binary: just repeatedly divide by 2 and get the remainder
    # the binary is the string that combines all the remainders tgt (last remainder -> first remainder)
    return bin_str(n//2) + bin_str(n%2)
```

```python
assert bin_str(0) == '0'
assert bin_str(1) == '1'
assert bin_str(2) == '10'
assert bin_str(3) == '11'
assert bin_str(5) == '101'
assert bin_str(8) == '1000'
assert bin_str(10) == '1010'
```
