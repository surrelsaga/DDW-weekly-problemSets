**CS5.** *Splitting Data Randomly:* Create a function to split the Data Frame randomly. The function should have the following arguments:
- `array_features`: which is the numpy array for the features.
- `array_target`: which is the numpy array for the target.
- `random_state`: which is the seed used to split randomly.
- `test_size`: which is the fraction for the test data set (0 to 1), by default is set to 0.5

The output of the function is a tuple of four items:
- `array_features_train`: which is the train set for the features 
- `array_features_test`: which is the test set for the features 
- `array_target_train`: which is the train set for the target 
- `array_target_test`: which is the test set for the target 

Hint:
- Use [`numpy.random.choice()`](https://docs.scipy.org/doc/numpy-1.13.0/reference/generated/numpy.random.choice.html) to select the indices for the test data set.
- Use [`pandas.DataFrame.drop()`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.drop.html) to drop the indices in the test data set to obtain the train data set.


```python
def split_data(array_features: np.ndarray, array_target: np.ndarray, 
               random_state: Optional[int]=None, 
               test_size: float=0.5) -> tuple[np.ndarray, np.ndarray, np.ndarray, np.ndarray]:
    # intialize randomizing seed
    np.random.seed(random_state)
    
    # GENERATE THE DATA SET
    
    # from doc
    # first arg: if pass in an int a, it will generate a 1D array by using np.arrange(a)
    # first arg: if pass in an ndarray (MUST BE 1D), a random sameple will be generated from its element
    
    # decision: pass in an int of the size of the data, np.random will have to generate random data in that size
    # array_features is 2D => .shape will return (no_of_rows, no_of_columns) => need size of rows => .shape[0]
    
    # replace=False is to random without replacement so we don't get duplicates
    
    # (supposedly we generate test data for these 2 arrays)
    # but they have the same rows, and when we remove them to get the remaining for training, it has to be the exact same rows
    # => only random once
    # array_feature_test = np.random.choice(array_features.shape[0], size = int(test_size * len(array_features)), replace=False)
    # array_target_test = np.random.choice(array_target.shape[0], size= int(test_size * len(array_target)), replace=False)
    
    # get the indexs of rows that are used to randomize dataset
    test_idx = np.random.choice(array_features.shape[0], size = int(test_size * len(array_features)), replace=False)
    
    
    # convert to pd.DataFrame to use drop()
    df_f = pd.DataFrame(array_features)
    df_t = pd.DataFrame(array_target)
    
    # All add .to_numpy since the .loc and .drop() only works on dataframe
    # after that convert back to ndarray using to_numpy()
    
    # need return those rows from the original arrays
    array_feature_test = df_f.loc[test_idx, :].to_numpy()
    array_target_test = df_t.loc[test_idx, :].to_numpy()
    
    # take all the remaining data (indices inside the array) and use them as training data
    array_feature_train = df_f.drop(test_idx).to_numpy()
    array_target_train = df_t.drop(test_idx).to_numpy()
    
    return array_feature_train, array_feature_test, array_target_train, array_target_test
    
```

```python
output: tuple[np.ndarray, np.ndarray, np.ndarray, np.ndarray] = split_data(df_features.to_numpy(), df_target.to_numpy(), random_state=100, test_size=0.3)
array_features_train, array_features_test, array_target_train, array_target_test = output

assert np.all(array_features_train.shape[0] == 355)
assert np.all(array_features_test.shape[0] == 151)
assert np.all(array_target_train.shape[0] == 355)
assert np.all(array_target_test.shape[0] == 151)

assert np.isclose(np.mean(array_features_train[:, 0]), 6.2968)
assert np.isclose(np.std(array_features_test[:, 1]), 2.22947)
assert np.isclose(np.median(array_target_train), 21.40)
assert np.isclose(np.median(array_target_test), 20.90)
```

**CS6.** *Standardization:* Write a function that takes in **two dimensional numpy array** where each of the column is one feature  in the data. The function should normalize each column according to the formula below. The function should also take in two optional arguments containing the means of each column and the standard deviation for each column. If these lists are not provided, the means and the standard deviation are computed from the input data.

$$normalized = \frac{data - \mu}{\sigma}$$

where $\mu$ is the mean of the data and $\sigma$ is the standard deviation of the data. **You need to normalize for each column respectively**. The function should return a new **two dimensional numpy array** as well as two **two dimensional numpy array** of mean and standard deviation that is used to normalized each column.

Use the following functions from Numpy:
- `np.mean(array, axis=0)`: This is to calculate the mean along the row axis (vertical).
- `np.std(array, axis=0)`: This is to calculate the standard deviation along the row axis (vertical).

Note: 
- The output means and standard deviation is a row vector. This means that it is a two dimensional array with a single row.
- Use `reshape(1,-1)` to create a row vector.
- Your function should be able to take in a single data point given as numpy array.

```python
def normalize_z(array: np.ndarray, columns_means: Optional[np.ndarray]=None, 
                columns_stds: Optional[np.ndarray]=None) -> tuple[np.ndarray, np.ndarray, np.ndarray]:
    assert columns_means is None or columns_means.shape == (1, array.shape[1])
    assert columns_stds is None or columns_stds.shape == (1, array.shape[1])
    
    if columns_means is None:
        # they want output to be "row vector" so use reshape(1, -1)
        columns_means = np.mean(array, axis=0).reshape(1,-1)
    if columns_stds is None:
        columns_stds = np.std(array, axis=0).reshape(1, -1)
        
    out = (array - columns_means) / columns_stds
    
    
    assert out.shape == array.shape
    assert columns_means.shape == (1, array.shape[1])
    assert columns_stds.shape == (1, array.shape[1])
    return out, columns_means, columns_stds
```

```python
features_array: np.ndarray = df_features.to_numpy()
norm_out: tuple[np.ndarray, np.ndarray, np.ndarray] = normalize_z(features_array)
data_norm, columns_means, columns_stds = norm_out
print(data_norm,columns_means, columns_stds)
data_df: pd.DataFrame = pd.DataFrame(data_norm, columns=df_features.columns)
stats: pd.DataFrame = data_df.describe()
display(stats)
assert isinstance(data_norm, np.ndarray)
assert np.isclose(stats.loc["mean", "RM"], 0.0) and \
       np.isclose(stats.loc["std", "RM"], 1.0, atol=1e-3)
assert np.isclose(stats.loc["mean", "DIS"], 0.0) and \
       np.isclose(stats.loc["std", "DIS"], 1.0, atol=1e-3)
assert np.isclose(stats.loc["mean", "INDUS"], 0.0) and \
       np.isclose(stats.loc["std", "INDUS"], 1.0, atol=1e-3)
assert np.isclose(columns_means[0][0], 6.2846)
assert np.isclose(columns_stds[0][2], 6.8536)
```
