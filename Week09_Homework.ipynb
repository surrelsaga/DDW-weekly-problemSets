**HW2.** *Count Plot:* Create a function to count how many records are diagnosed as Malignant and Benign. The function should return a tuple: `(n_malignant, n_benign)`, where `n_malignant` is the number of records diagnosed as Malignant cell and `n_benign` is the number of records diagnosed as Benign.

Use Count plot to verify the answer.

Reference:
- [Count Plot](https://seaborn.pydata.org/generated/seaborn.countplot.html)


```python
def count_cell(target: np.ndarray) -> tuple[int, int]:
    n_malignant: int = 0
    n_benign: int = 0
    
    # numpy array is iterable just like normal array
    # but small difference
    # for e.g:
    # [1, 2, 3] * 2 = [1, 2, 3, 1, 2, 3]
    # np.array([1,2,3]) * 2 = [2, 4, 6]
    
    for value in target: 
        if value == 'M':
            n_malignant += 1
        else:
            n_benign += 1
    
    return (n_malignant, n_benign)
        
```

**HW3.** *Normalization:* Create a function that takes in **two dimensional numpy array** as the input and returns the normalized numpy array as the output. **Each column** is normalized separately using **min-max normalization**. The function should return a new two dimensional numpy array instead of modifying the input array. The function should also take in two optional arguments containing the minimums and the maximus of each column. If these lists are not provided, the minimums and the maximums are computed from the input data frame.

$$normalized = \frac{data - min}{max - min}$$

Use the following functions from Numpy:
- `np.copy()`: This is to create a new copy of the data frame.
- `np.min(array, axis=0)`: This is to get the minimum along the index axis.
- `np.max(array, axis=0)`: This is to get the maximum along the index axis.

Note: 
- Your function should be able to take in a single data point given as a numpy array.

```python
def normalize_minmax(array_in: np.ndarray, columns_mins: Optional[np.ndarray]=None, 
                     columns_maxs: Optional[np.ndarray]=None) -> tuple[np.ndarray, np.ndarray, np.ndarray]:
    assert columns_mins is None or columns_mins.shape == (1, array_in.shape[1])
    assert columns_maxs is None or columns_maxs.shape == (1, array_in.shape[1])
    
    array_copy = np.copy(array_in)
    
    # extra arg: keepdims cuz the return is (6,..)
    # cuz the return value is an np array with dimention (6,..)
    # 6 min/max from 6 columns (min of each column)
    # they don't have a size yet, we just want them to be inside one row and spread out
    # => use keepdims=True to get (1, 6)
    if columns_mins is None:
        columns_mins = np.min(array_copy, axis=0, keepdims=True)
    if columns_maxs is None:
        columns_maxs = np.max(array_copy, axis=0, keepdims=True)
    
    out = (array_copy - columns_mins) / (columns_maxs - columns_mins)
    
    assert out.shape == array_in.shape
    assert columns_mins.shape == (1, array_in.shape[1])
    assert columns_maxs.shape == (1, array_in.shape[1])
    return out, columns_mins, columns_maxs
```

```python
feature_array: np.ndarray = df_features.to_numpy()
norm_out: tuple[np.ndarray, np.ndarray, np.ndarray] = normalize_minmax(feature_array)
data_norm, columns_mins, columns_maxs = norm_out
print(columns_mins, columns_maxs)
data_df: pd.DataFrame = pd.DataFrame(data_norm, columns=df_features.columns)
stats: pd.DataFrame = data_df.describe()
display(stats)
assert isinstance(data_norm, np.ndarray)
assert stats.loc["max", "radius_mean"] == 1.0 and \
       stats.loc["min", "radius_mean"] == 0 and \
       np.isclose(stats.loc["mean", "radius_mean"], 0.338222)
assert stats.loc["max", "texture_mean"] == 1.0 and \
       stats.loc["min", "texture_mean"] == 0 and\
       np.isclose(stats.loc["mean", "texture_mean"], 0.323965)
assert np.isclose(columns_mins[0][0], 6.981)
assert np.isclose(columns_mins[0][2], 43.79)
assert np.isclose(columns_mins[0][4], 0.05263)
assert np.isclose(columns_maxs[0][0], 28.11)
assert np.isclose(columns_maxs[0][2], 188.5)
assert np.isclose(columns_maxs[0][4], 0.1634)
```

**HW4.** *Splitting the Data:* Use the function to split the breast cancer data set into a training data set and a testing data set. Use `random_state=100` and `test_size=0.3` and the normalized features data set from the previous exercise.

```python
# copy your answer from the cohort problem

def split_data(array_features: np.ndarray, array_target: np.ndarray, 
               random_state: Optional[int]=None, 
               test_size: float=0.5) -> tuple[np.ndarray, np.ndarray, np.ndarray, np.ndarray]:
    
    # step 1: initialize the seed to random
    np.random.seed(random_state)
    
    # PHASE: PREPARE THE TESTING DATA
    
    # fraction * total len = size used for testing
    size_to_test = int(test_size * len(array_features))
    
    # array_features.shape[0]: 2D-array so .shape[0] => to get the total number of rows
    # replace=False, to avoid duplicate, like in probability and stats, no replacement
    test_indexes = np.random.choice(array_features.shape[0], size=size_to_test, replace=False)
    
    
    # PHASE: extract the data for testing and and the remainings are used for training
    # use .loc() and .drop() => need type pd.DataFrame
    
    # convert to DataFrame
    df_f = pd.DataFrame(array_features)
    df_t = pd.DataFrame(array_target)
    
    # extract testing data + to_numpy since the output must be in type np array
    array_feature_test = df_f.loc[test_indexes, :].to_numpy()
    array_target_test = df_t.loc[test_indexes, :].to_numpy()
    
    # the remaining are used for training + to_numpy since the output must be in type np array
    array_feature_train = df_f.drop(test_indexes).to_numpy()
    array_target_train = df_t.drop(test_indexes).to_numpy()
    
    return array_feature_train, array_feature_test, array_target_train, array_target_test
    

 
```

```python
# call split_data() with seed of 100 and test size of 30%
datasets_tuple = split_data(df_features.to_numpy(), df_target.to_numpy(), random_state=100, test_size=0.3)

# extract features data
array_feature_train, array_feature_test, _, _ = datasets_tuple

# normalize the features data
norm_out_train: tuple[np.ndarray, np.ndarray, np.ndarray] = normalize_minmax(array_feature_train)
array_feature_norm_train, means, stds = norm_out_train

# use the means and stds from the training set to normalize the test set
norm_out_test: tuple[np.ndarray, np.ndarray, np.ndarray] = normalize_minmax(array_feature_test, means, stds)
array_feature_norm_test, _, _ = norm_out_test
```

```python
df_features_train: pd.DataFrame = pd.DataFrame(array_feature_norm_train, columns=df_features.columns)
df_features_test: pd.DataFrame = pd.DataFrame(array_feature_norm_test, columns=df_features.columns)

display(df_features_train.describe())
display(df_features_test.describe())
```

```python
assert isinstance(array_feature_norm_train, np.ndarray)
assert isinstance(array_feature_norm_test, np.ndarray)

assert array_feature_norm_train.shape == (399, 6)
assert array_feature_norm_test.shape == (170, 6)

assert np.isclose(array_feature_norm_train.mean().mean(), 0.29585, rtol=1e-3) 
assert np.isclose(array_feature_norm_test.mean().mean(), 0.309412, rtol=1e-3) 
```
