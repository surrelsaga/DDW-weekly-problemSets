Q1. Design a class named `Rectangle` to represent a rectangle. The class contains:
* two properties called `width` and `height`.
* the initializer should allow the object creation specifying the width and the height. The default values are 1 and 2 for the width and the height respectively. 
* a method named `get_area()` that returns the area of this rectangle.
* a method named `get_perimeter()` that returns the perimeter.
* the setter for the properties must ensure that zero or negative number should never be set into the `width` and `height` of the rectangle.

```python
class Rectangle:
    def __init__(self, width=1, height=2):
        self._width = width
        self._height = height
    
    @property
    def width(self):
        return self._width
    
    @width.setter
    def width(self, val):
        if val > 0:
            self._width = val
    
    @property
    def height(self):
        return self._height
    
    @height.setter
    def height(self, val):
        if val > 0:
            self._height = val
    
    def get_area(self):
        return self._width * self._height
    
    def get_perimeter(self):
        return 2 * (self._width + self._height)
```

```python
r1 = Rectangle(4, 40)
r2 = Rectangle(3.5, 35.7)

for r in (r1, r2):
    print(r.width, r.height, r.get_area(), r.get_perimeter())
```

Q3. Create a class called `Account` to represent a Bank account. Its initialization takes in an initial balance of the account when it is opened. By default, it should be initialized to 0.0. The class has three methods:
- `balance()` that returns the balance of the account.
- `deposit(amount)` that deposit `amount` into the balance.
- `credit_limit()` that returns the credit limit of the account. The credit limit is the minimum of either half of the balance or 1,000,000. 

```python
class Account:
    def __init__(self, balance=0.0):
        self._balance = balance

    def balance(self):
        return self._balance
    
    def deposit(self, amount):
        self._balance += amount
    
    def credit_limit(self):
        return min(self._balance // 2, 1000000)
```

Q4. Create a class called `Time` to represent time. Its initialization can take two arguments, i.e. `hours` and `minutes`. By default they should be set to zero. We do not represent the seconds in this class. 

Implement the following methods:
- `__add__(self, other)` to add two Time object instances.
- `__str__(self)` to return a string in this format `HH:MM`. 

```python
class Time:
    def __init__(self, hours=0, minutes=0):
        self._hours = hours
        self._minutes = minutes
    
    def __add__(self, other):
        total_min = (self._hours * 60 + self._minutes) + (other._hours * 60 + other._minutes)
        
        # because we don't use AM or PM
        total_min = total_min % (24 * 60)
        
        hours = total_min // 60
        minutes = total_min % 60
        
        return Time(hours, minutes)
    
    def __str__(self):
        output = ""
        if self._hours < 10:
            output += f"0{self._hours}:"
        else:
            output += f"{self._hours}:"
        
        if self._minutes < 10:
            output += f"0{self._minutes}"
        else:
            output += f"{self._minutes}"
            
        return output
```

```python
t1: Time = Time(6, 30)
t2: Time = Time(minutes=45)
t3: Time = Time(23, 59)

assert str(t1) == "06:30"
assert str(t2) == "00:45"
assert str(t3) == "23:59"

out: Time = t1 + t2
print(out)
assert str(out) == "07:15"
out: Time = t3 + t2
print(out)
assert str(out) == "00:44"
out: Time = t1 + t2 + t3
assert str(out) == "07:14"
```

Q5. Create a class called `SignFraction` which can handle both positive and negative fraction. It has an additional **computed property** called `negative` which is a Boolean. Use the calss `Fraction` in your cohort and homework and modify all the methods so that it works with this new class.

```python
def gcd(a: int, b: int) -> int:
    if b == 0:
        return a
    else:
        return gcd(b, a % b)


class SignFraction:
    def __init__(self, num: int, den: int) -> None:
        assert isinstance(num, int)
        assert isinstance(den, int)
        self._num = num
        self._den = 1 if den == 0 else den
        # keep the sign on the numerator, denominator always positive
        if self._den < 0:
            self._num = -self._num
            self._den = -self._den

    
    @property
    def num(self) -> int:
        return abs(self._num)
    
    @num.setter
    def num(self, val: int) -> None:
        if isinstance(val, (int, float)):
            self._num = int(val)
    
    @property
    def den(self) -> int:
        return self._den
    
    @den.setter
    def den(self, val: int) -> None:
        if not isinstance(val, (int, float)):
            return

        val = int(val)
        self._den = 1 if val == 0 else val
        if self._den < 0:
            self._num = -self._num
            self._den = -self._den
    
    @property
    def negative(self) -> bool:
        return self._num < 0

    def __str__(self) -> str:
        if self.negative:
            return f"-{self.num}/{self.den}"
        
        return f"{self.num}/{self.den}"
    
    
    def simplify(self) -> SignFraction:
        common_divisor = gcd(abs(self._num), abs(self._den))
        if common_divisor == 0:
            return SignFraction(self._num, self._den)

        new_num = self._num // common_divisor
        new_den = self._den // common_divisor

        return SignFraction(new_num, new_den)
    
    def __add__(self, other) -> SignFraction:
        result_num = self._num * other._den + self._den * other._num
        result_den = self._den * other._den

        return SignFraction(result_num, result_den).simplify()
        
    def __sub__(self, other) -> SignFraction:
        result_num = self._num * other._den - self._den * other._num
        result_den = self._den * other._den

        return SignFraction(result_num, result_den).simplify()
    
    def __mul__(self, other) -> SignFraction:
        result_num = self._num * other._num
        result_den = self._den * other._den

        return SignFraction(result_num, result_den).simplify()
    
    def __eq__(self, other) -> bool:
        simplified_self = self.simplify()
        simplified_other = other.simplify()

        return (
            simplified_self._num == simplified_other._num
            and simplified_self._den == simplified_other._den
        )
    
    def __lt__(self, other) -> bool:
        return self._num * other._den < other._num * self._den
    
    def __le__(self, other) -> bool:
        return self < other or self == other
    
    def __gt__(self, other) -> bool:
        return other < self
    
    def __ge__(self, other) -> bool:
        return other <= self
    
```

Q5. Create a class called `Node` which we will use later on to represent Binary Search Tree. This class contains:
- a `left` node property and a `right` node property
- a `parent` node property
- the setter for these properties must ensure that only `Node` instance or a `None` object can be set as values to these properties. 
- a `value` property that can be used to store data. Value cannot be a `None` object.
- Create an initializer that initializes a `Node` object instance with a value and set all the three properties. The value has to be given while the left, right and parent properties have a default value of `None`.   

```python
class Node:
    def __init__(self, value: Any, parent: Optional[Node]=None, left: Optional[Node]=None, right: Optional[Node]=None) -> None:
        assert value is not None, "Value cannot be None"
        self._value = value
        self._parent = parent
        self._left = left
        self._right = right

    @property
    def value(self) -> Any:
        return self._value
    
    @value.setter
    def value(self, value: Any) -> None:
        if value is not None:
            self._value = value

    @property
    def left(self) -> Optional[Node]:
        return self._left
    
    @left.setter
    def left(self, left: Optional[Node]) -> None:
        if isinstance(left, Node) or left is None:
            self._left = left

    @property
    def right(self) -> Optional[Node]:
        return self._right
    
    @right.setter
    def right(self, right: Optional[Node]) -> None:
        if isinstance(right, Node) or right is None:
            self._right = right
        
    @property
    def parent(self) -> Optional[Node]:
        return self._parent
    
    @parent.setter
    def parent(self, parent: Optional[Node]) -> None:
        if isinstance(parent, Node) or parent is None:
            self._parent = parent
```

```python
root: Node = Node(8)
assert root.value == 8 
assert root.left == None and root.right == None and root.parent == None

n3: Node = Node(3, root)
assert n3.value == 3
assert n3.left == None and n3.right == None and n3.parent == root

# value cannot be None
n3.value = None
assert n3.value == 3
n3.value = 10
assert n3.value == 10
n3.value = 3

# testing left setter
root.left = n3
assert root.left == n3

# testing right setter
n10: Node = Node(10, root)
root.right = n10
assert root.right == n10

# testing parent setter
assert n10.parent == root
```

Q6. Create a class called `BST` to represent Binary Search Tree. This class contains:
* a property which is named `root`. This property can only either a `Node` object or a None. 
* a computed property called `size` which returns the number of nodes in the BST. 
* an initializer which takes nothing.
* a method named `add(value)` to add a value to the BST. 
 

```python
class BST:
    def __init__(self) -> None:
        self._size = 0
        self._root = None

    @property
    def size(self) -> int:
        return self._size
    
    @property
    def root(self) -> Optional[Node]:
        return self._root
    
    @root.setter
    def root(self, root: Optional[Node]) -> None:
        if root is None or isinstance(root, Node):
            self._root = root

    def add(self, value: Any) -> None:
        if self.root is None:
            self.root = Node(value)
        # Add like a binary tree (heap is one type)
        else:
            current = self.root
            # Binary search tree rule: in one node, left child must be smaller, right child must be larger than the value of the node
            while True:
                # this case is only for adding left
                # if there is already left (children) of the node, but have to stick to the rule, can not add to the right
                # we go down to that left children, add as the left children of that left children
                if value < current.value:
                    # meaning there hasn't a left child yet, then create one
                    if current.left is None:
                        current.left = Node(value)
                        # after added then just end the loop
                        break
                    else:
                        current = current.left
                else:
                    # meaning there hasn't a right child yet, then create one
                    if current.right is None:
                        current.right = Node(value)
                        
                        break
                    else:
                        current = current.right
        self._size += 1


```

Q8. Add a method `search(key)` to find if `key` is in the BST. Returns the key if it is found or return `None` if it is not found.

```python
class BST:
    def __init__(self, array: list[Any]):
        self._root = None
        self._size = 0
        for value in array:
            self.add(value)

    @property
    def size(self) -> int:
        return self._size
    
    @property
    def root(self) -> Optional[Node]:
        return self._root
    
    @root.setter
    def root(self, root: Optional[Node]) -> None:
        if root is None or isinstance(root, Node):
            self._root = root

    def add(self, value: Any) -> None:
        if self.root is None:
            self.root = Node(value)
        # Add like a binary tree (heap is one type)
        else:
            current = self.root
            # Binary search tree rule: in one node, left child must be smaller, right child must be larger than the value of the node
            while True:
                # this case is only for adding left
                # if there is already left (children) of the node, but have to stick to the rule, can not add to the right
                # we go down to that left children, add as the left children of that left children
                if value < current.value:
                    # meaning there hasn't a left child yet, then create one
                    if current.left is None:
                        current.left = Node(value)
                        # after added then just end the loop
                        break
                    else:
                        current = current.left
                else:
                    # meaning there hasn't a right child yet, then create one
                    if current.right is None:
                        current.right = Node(value)
                        
                        break
                    else:
                        current = current.right
        self._size += 1

    
    def search(self, key: Any) -> Optional[Node]:
        current = self.root
        
        while True:
            if current is None:
                return None
            
            if current.value == key:
                return current
            
            # Easy search: recursion
            if key < current.value:
                current = current.left
            else:
                current = current.right

```

* Items are pushed into a priority queue with a numeric score, called a *cost*.
* When it is time to pop an item, the item in the priority queue with the least cost is returned and removed from the priority queue. 

```python
class PriorityQueue:

    def __init__(self) -> None:
        self.queue: list[Any] = []

    def push(self, item: Any, cost: Number) -> None:
        self.queue.append( (cost, item) )

    def pop(self) -> Any:
        if self.is_empty:
            return None
        
        return self.queue.pop( self.find_min_cost_index() )
                

    def peek(self) -> Any:
        if self.is_empty:
            return None 
        
        return self.queue[self.find_min_cost_index()]

    def find_min_cost_index(self) -> Optional[int]:
        if self.is_empty:
            return None
        
        # # Let the first cost of first item be the thing to compare
        # min_cost = self.queue[0][0]
        # for item in self.queue:
        #     if item[0] <= min_cost:
        #         min_cost = item[0]
        
        # for i in range(len(self.queue)):
        #     if self.queue[i][0] == min_cost:
        #         return i
                # Let the first cost of first item be the thing to compare
        
        
        # one loop only
        min_index = 0
        for i in range(len(self.queue)):
            if self.queue[i][0] < self.queue[min_index][0]:
                min_index = i
        
        return min_index

    @property
    def is_empty(self) -> bool:
        return len(self.queue) == 0

    @property
    def size(self) -> int:
        return len(self.queue)
```
